```python
import pandas as pd
df_clientes = pd.read_csv('/Users/saraelbachouti/Desktop/UCV-Churn/data/raw/clientes.csv')
df_churn = pd.read_csv('data/raw/churn_target.csv')
```

```python
df_churn.info()
df_churn.isnull().sum()
df_churn.describe()
```

```python
df_churn.info()
df_churn["churn"].value_counts()
```

```python
df_churn_final = df_churn.groupby("cliente_id")["churn"].max().reset_index()
```

```python
df_clientes = df_clientes.drop_duplicates()
```

```python
df_clientes.isnull().sum()
```

```python
df_clientes["edad"] = df_clientes["edad"].fillna(df_clientes["edad"].median())
df_clientes["ingreso_estimado"] = df_clientes["ingreso_estimado"].fillna(df_clientes["ingreso_estimado"].median())
df_clientes["antiguedad_meses"] = df_clientes["antiguedad_meses"].fillna(df_clientes["antiguedad_meses"].median())

df_clientes["sexo"] = df_clientes["sexo"].fillna("desconocido")
df_clientes["estado_civil"] = df_clientes["estado_civil"].fillna("desconocido")
```

```python
df_clientes["tipo_plan"] = df_clientes["tipo_plan"].str.lower().str.strip()
df_clientes["region"] = df_clientes["region"].str.lower().str.strip()
df_clientes["tipo_zona"] = df_clientes["tipo_zona"].str.lower().str.strip()
```

```python
df_clientes["edad"].describe()
```

```python
df_clientes.to_csv('data/processed/clientes_cleaned.csv', index=False)
```

```python
df_churn_final.to_csv('data/processed/churn_final_cleaned.csv', index=False)
```
